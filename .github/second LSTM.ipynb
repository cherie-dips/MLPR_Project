```python
# Import necessary libraries if not already imported
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, Flatten, Dense, TimeDistributed, LSTM, Lambda, Average

# Define input shape based on your data
# (sequence_length, height, width, channels)
input_shape = (11, 128, 128, 3)  # Adjust if your dimensions are different
num_classes = 4  # For pH values 5,6,7,8

# Build the hybrid model
inp = Input(shape=input_shape)

# 1. Process each timepoint individually 
x = TimeDistributed(Conv2D(32, (3, 3), activation='relu'))(inp)
x = TimeDistributed(MaxPooling2D())(x)
x = TimeDistributed(Conv2D(64, (3, 3), activation='relu'))(x)
x = TimeDistributed(MaxPooling2D())(x)
x = TimeDistributed(Flatten())(x)

# Individual timepoint predictions
individual_preds = TimeDistributed(Dense(num_classes, activation='softmax'))(x)

# 2. Process the sequence
sequence_features = LSTM(128)(x)
sequence_preds = Dense(num_classes, activation='softmax')(sequence_features)

# Combine both predictions (from last timepoint and from sequence)
final_preds = Average()([individual_preds[:,-1,:], sequence_preds])

# Create and compile the model
hybrid_model = Model(inputs=inp, outputs=final_preds)
hybrid_model.compile(optimizer='adam', 
                    loss='categorical_crossentropy', 
                    metrics=['accuracy'])

# Display the model summary
hybrid_model.summary()

# Train the model using your existing generators
history = hybrid_model.fit(train_gen, 
                          validation_data=val_gen, 
                          epochs=20,  # Increase epochs
                          verbose=1)

# Plot training history
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'], loc='upper left')

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'], loc='upper left')

plt.tight_layout()
plt.show()

# Evaluate on validation set
val_loss, val_accuracy = hybrid_model.evaluate(val_gen)
print(f"Validation Accuracy: {val_accuracy:.4f}")
```

```python
# Visualize samples from your data
def visualize_sequence(generator, sample_index=0):
    """Visualize an image sequence from the generator"""
    # Get one batch
    batch_x, batch_y = generator[0]
    
    # Extract one sequence
    sequence = batch_x[sample_index]
    label = np.argmax(batch_y[sample_index]) + 5  # Convert back to actual pH
    
    # Plot the sequence
    fig, axes = plt.subplots(1, sequence.shape[0], figsize=(20, 4))
    for i, ax in enumerate(axes):
        ax.imshow(sequence[i])
        ax.set_title(f"Day {i+1}")
        ax.axis('off')
    plt.suptitle(f"pH: {label}")
    plt.tight_layout()
    plt.show()
    
    # Print image statistics to check for preprocessing issues
    print(f"Image shape: {sequence[0].shape}")
    print(f"Value range: {sequence[0].min()} - {sequence[0].max()}")
    
    return sequence

# Visualize a sample
sample_sequence = visualize_sequence(train_gen)
```

```python
def file_path_extractor(idx, df, day, main_dir):
    """Fixed function to extract file paths correctly"""
    row = df[df['Well'] == idx]
    if row.empty:
        print(f"Well {idx} not found in dataframe")
        return None
    
    pH = int(row['pH'].values[0])
    img_name = row[day].values[0]
    
    if pd.isna(img_name) or img_name == "":
        print(f"Missing image name for Well {idx}, {day}")
        return None
    
    hour = day_to_hour[day]
    folder = f"{main_dir}/{hour} hr/pH{pH} Hydrolytic"
    full_path = os.path.join(folder, img_name)
    
    # Additional validation
    if not os.path.exists(folder):
        print(f"Folder doesn't exist: {folder}")
        return None
    
    if not os.path.exists(full_path):
        print(f"Image doesn't exist: {full_path}")
        return None
        
    return full_path
```

```python
def build_hybrid_model(input_shape, num_classes=4):
    # CNN branch for individual image classification
    inp = Input(shape=input_shape)
    
    # 1. Process each timepoint individually 
    x = TimeDistributed(Conv2D(32, (3, 3), activation='relu'))(inp)
    x = TimeDistributed(MaxPooling2D())(x)
    x = TimeDistributed(Conv2D(64, (3, 3), activation='relu'))(x)
    x = TimeDistributed(MaxPooling2D())(x)
    x = TimeDistributed(Flatten())(x)
    
    # Individual timepoint predictions
    individual_preds = TimeDistributed(Dense(num_classes, activation='softmax'))(x)
    
    # 2. Process the sequence
    sequence_features = LSTM(128)(x)
    sequence_preds = Dense(num_classes, activation='softmax')(sequence_features)
    
    # Combine both predictions
    final_preds = tf.keras.layers.Average()([individual_preds[:,-1,:], sequence_preds])
    
    return Model(inputs=inp, outputs=final_preds)
```

```python
def build_feature_enhanced_model(img_shape, feature_shape, num_classes=4):
    # Image input branch
    img_input = Input(shape=img_shape)
    x1 = TimeDistributed(Conv2D(32, (3, 3), activation='relu'))(img_input)
    x1 = TimeDistributed(MaxPooling2D())(x1)
    x1 = TimeDistributed(Conv2D(64, (3, 3), activation='relu'))(x1)
    x1 = TimeDistributed(MaxPooling2D())(x1)
    x1 = TimeDistributed(Flatten())(x1)
    x1 = LSTM(64)(x1)
    
    # Feature input branch
    feature_input = Input(shape=feature_shape)
    x2 = LSTM(32)(feature_input)
    
    # Combine both branches
    combined = tf.keras.layers.concatenate([x1, x2])
    x = Dense(64, activation='relu')(combined)
    output = Dense(num_classes, activation='softmax')(x)
    
    return Model(inputs=[img_input, feature_input], outputs=output)
```

```python
# Simple model using only the last timepoint
def build_last_day_model(input_shape, num_classes=4):
    inp = Input(shape=input_shape)
    
    # Extract only the last day (index -1)
    last_day = Lambda(lambda x: x[:, -1, :, :, :])(inp)
    
    # Process the last day only
    x = Conv2D(32, (3, 3), activation='relu')(last_day)
    x = MaxPooling2D()(x)
    x = Conv2D(64, (3, 3), activation='relu')(x)
    x = MaxPooling2D()(x)
    x = Flatten()(x)
    x = Dense(128, activation='relu')(x)
    out = Dense(num_classes, activation='softmax')(x)
    
    return Model(inputs=inp, outputs=out)
```

```python
import os
import numpy as np
import pandas as pd
import cv2
from tensorflow.keras.utils import Sequence
from sklearn.model_selection import train_test_split
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, Flatten, Dense, TimeDistributed, LSTM

# Mapping day labels to hours
day_to_hour = {
    'Day 1': 0, 'Day 2': 24, 'Day 3': 30, 'Day 4': 48, 'Day 5': 72,
    'Day 6': 95, 'Day 7': 120, 'Day 8': 168, 'Day 9': 192,
    'Day 10': 216, 'Day 11': 264
}
timestamp_to_class = {v: i for i, v in enumerate(day_to_hour.values())}

# âœ… Fix: function to extract path
def file_path_extractor(idx, df, day, main_dir):
    row = df[df['Well'] == idx]
    if row.empty:
        return None
    pH = int(row['pH'].values[0])
    img_name = row[day].values[0]
    hour = day_to_hour[day]
    folder = f"{main_dir}/{hour} hr/pH{pH} Hydrolytic"
    return os.path.join(folder, img_name)

# âœ… Data generator to load sequences of images per sample
class ImageSequenceGenerator(Sequence):
    def __init__(self, df, main_dir, batch_size=8, target_size=(128, 128), shuffle=True):
        self.df = df
        self.batch_size = batch_size
        self.target_size = target_size
        self.main_dir = main_dir
        self.shuffle = shuffle
        self.indexes = df['Well'].tolist()
        self.days = list(day_to_hour.keys())
        self.on_epoch_end()

    def __len__(self):
        return int(np.floor(len(self.indexes) / self.batch_size))

    def __getitem__(self, index):
        batch_ids = self.indexes[index*self.batch_size:(index+1)*self.batch_size]
        return self.__data_generation(batch_ids)

    def on_epoch_end(self):
        if self.shuffle:
            np.random.shuffle(self.indexes)

    def __data_generation(self, batch_ids):
        X, y = [], []
        for idx in batch_ids:
            seq_imgs = []
            for day in self.days:
                path = file_path_extractor(idx, self.df, day, self.main_dir)
                if path and os.path.exists(path):
                    img = cv2.imread(path)
                    img = cv2.resize(img, self.target_size)
                    img = img / 255.0
                else:
                    img = np.zeros((*self.target_size, 3))  # fallback blank
                seq_imgs.append(img)
            X.append(seq_imgs)
            label = self.df[self.df['Well'] == idx]['pH'].values[0] - 5  # assuming pH âˆˆ {5,6,7,8}
            y.append(label)
        return np.array(X), tf.keras.utils.to_categorical(np.array(y), num_classes=4)

# âœ… CNN-LSTM Model
def build_model(input_shape, num_classes=4):
    inp = Input(shape=input_shape)  # (sequence_len, height, width, channels)
    x = TimeDistributed(Conv2D(16, (3, 3), activation='relu'))(inp)
    x = TimeDistributed(MaxPooling2D((2, 2)))(x)
    x = TimeDistributed(Conv2D(32, (3, 3), activation='relu'))(x)
    x = TimeDistributed(MaxPooling2D((2, 2)))(x)
    x = TimeDistributed(Flatten())(x)
    x = LSTM(64)(x)
    out = Dense(num_classes, activation='softmax')(x)
    return Model(inputs=inp, outputs=out)

# âœ… Load and split data
main_dir = 'New MLPR Data'
train_df, val_df = train_test_split(train, test_size=0.2, random_state=42)

train_gen = ImageSequenceGenerator(train_df, main_dir)
val_gen = ImageSequenceGenerator(val_df, main_dir)

# âœ… Compile and train model
model = build_model(input_shape=(11, 128, 128, 3))
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
model.summary()

# âœ… Training
model.fit(train_gen, validation_data=val_gen, epochs=10)
```
